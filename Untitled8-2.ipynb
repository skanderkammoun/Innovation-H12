```python
import fitz  # PyMuPDF

pdf_path = "page de garde volume I.pdf"  # Path to your PDF
output_folder = ""  # Folder to save extracted images

# Open the PDF
doc = fitz.open(pdf_path)

# Loop through pages and extract images
for page_number in range(len(doc)):
    for img_index, img in enumerate(doc[page_number].get_images(full=True)):
        xref = img[0]
        base_image = doc.extract_image(xref)
        image_bytes = base_image["image"]
        img_ext = base_image["ext"]

        # Save image
        img_filename = f"{output_folder}extracted_image_{page_number+1}_{img_index+1}.{img_ext}"
        with open(img_filename, "wb") as f:
            f.write(image_bytes)

print("Image extraction completed.")
```

```python
from diffusers import StableDiffusionInpaintPipeline
import torch
from PIL import Image, ImageDraw

# Load the model
pipe = StableDiffusionInpaintPipeline.from_pretrained("runwayml/stable-diffusion-inpainting")
pipe.to("cuda" if torch.cuda.is_available() else "cpu")

# Load the ruined image
image_path = "Image2.jpg"  # Make sure the file exists
image = Image.open(image_path).convert("RGB")

# Create a mask (white = areas to restore, black = keep original)
mask = Image.new("L", image.size, 0)  # Black background
draw = ImageDraw.Draw(mask)
draw.rectangle([50, 50, image.size[0] - 50, image.size[1] - 50], fill=255)  # White rectangle for inpainting

# Define the prompt for AI restoration
prompt = "Reconstruct this ruin to its original form, ancient Roman architecture, high detail, realistic lighting"

# Perform inpainting (restoring the image)
restored_image = pipe(prompt=prompt, image=image, mask_image=mask).images[0]

# Save and display the reconstructed image
restored_image.save("imageHneya.png")
restored_image.show()
```

```python
from diffusers import StableDiffusionInpaintPipeline
import torch
from PIL import Image, ImageDraw

# Load the model
pipe = StableDiffusionInpaintPipeline.from_pretrained("runwayml/stable-diffusion-inpainting")
pipe.to("cuda" if torch.cuda.is_available() else "cpu")

# Load the ruined image
image_path = "Image1.jpg"  # Make sure the file exists
image = Image.open(image_path).convert("RGB")

# Create a mask (white = areas to restore, black = keep original)
mask = Image.new("L", image.size, 0)  # Black background
draw = ImageDraw.Draw(mask)
draw.rectangle([50, 50, image.size[0] - 50, image.size[1] - 50], fill=255)  # White rectangle for inpainting

# Define the prompt for AI restoration
prompt = "Reconstruct the ancient nymphaeum of Sufes, accurately depicting its semi-circular structure with a three-level water flow system. Ensure that the water channels remain hidden, with the flow emerging from sculpted openings, forming cascading sheets for the largest outlets. Include classical Roman architectural elements such as columns, niches, and ornate facades, incorporating statues and decorative reliefs. Use historically accurate stone and brick materials, with realistic texturing and signs of natural weathering. The lighting should reflect the warm, natural environment of ancient Tunisia, casting soft shadows that enhance depth and realism. The final image should be highly detailed, resembling a high-resolution historical reconstruction, showcasing both the aesthetic grandeur and functional sophistication of this ancient Roman water system"
# Perform inpainting (restoring the image)
restored_image = pipe(prompt=prompt, image=image, mask_image=mask).images[0]

# Save and display the reconstructed image
restored_image.save("mimi2.png")
restored_image.show()
```

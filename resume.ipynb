```python
# Étape 1 : Installer les bibliothèques nécessaires
!pip install transformers
!pip install PyMuPDF
!pip install pdfplumber

# Import the fitz module from PyMuPDF
import fitz

# This line is unnecessary and should be removed
# import PyMuPDF

def extract_text_from_pdf(pdf_path):
    """
    Extrait le texte d'un fichier PDF using PyMuPDF.
    :param pdf_path: Chemin du fichier PDF.
    :return: Texte extrait.
    """
    text = ""
    with fitz.open(pdf_path) as pdf_file:  # Use fitz.open from PyMuPDF
        for page_num in range(pdf_file.page_count):
            page = pdf_file[page_num]
            text += page.get_text() + "\n"
    return text

# Extraire le texte du PDF using PyMuPDF
pdf_text = extract_text_from_pdf(pdf_path)
print(f"Texte extrait du PDF :\n{pdf_text[:1000]}...")  # Affiche les 1000 premiers caractères

# # Extraire le texte du PDF using PyPDF2
# pdf_text = extract_text_from_pdf(pdf_path)
# print(f"Texte extrait du PDF :\n{pdf_text[:1000]}...")  # Affiche les 1000 premiers caractères
# Étape 4 : Charger un modèle de résumé (BART ou T5) sur le GPU
# You need to import pipeline from transformers
from transformers import pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device=0 if torch.cuda.is_available() else -1)

# Étape 5 : Diviser le texte en morceaux (si le texte est trop long)
def split_text(text, max_length=1024):
    """
    Divise le texte en morceaux de taille maximale pour le modèle.
    :param text: Texte à diviser.
    :param max_length: Taille maximale d'un morceau.
    :return: Liste des morceaux de texte.
    """
    words = text.split()
    chunks = []
    current_chunk = []
    current_length = 0

    for word in words:
        if current_length + len(word) + 1 <= max_length:
            current_chunk.append(word)
            current_length += len(word) + 1
        else:
            chunks.append(" ".join(current_chunk))
            current_chunk = [word]
            current_length = len(word) + 1

    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks

# Diviser le texte en morceaux
text_chunks = split_text(pdf_text)
print(f"Nombre de morceaux de texte : {len(text_chunks)}")

# Étape 6 : Résumer chaque morceau de texte
def summarize_text(text_chunks, summarizer, max_length=130, min_length=30):
    """
    Résume chaque morceau de texte.
    :param text_chunks: Liste des morceaux de texte.
    :param summarizer: Modèle de résumé.
    :param max_length: Longueur maximale du résumé.
    :param min_length: Longueur minimale du résumé.
    :return: Liste des résumés.
    """
    summaries = []
    for chunk in text_chunks:
        summary = summarizer(chunk, max_length=max_length, min_length=min_length, do_sample=False)
        summaries.append(summary[0]['summary_text'])
    return summaries

# Résumer chaque morceau de texte
summaries = summarize_text(text_chunks, summarizer)
print(f"Résumés générés :\n{summaries}")

# Étape 7 : Combiner les résumés en un seul texte
final_summary = " ".join(summaries)
print(f"Résumé final :\n{final_summary}")

# Étape 8 : Sauvegarder le résumé dans un fichier
output_filename = "resume.txt"
with open(output_filename, "w", encoding="utf-8") as output_file:
    output_file.write(final_summary)

print(f"Le résumé a été sauvegardé dans {output_filename}")

# Télécharger le fichier texte (optionnel)
from google.colab import files
files.download(output_filename)
```
